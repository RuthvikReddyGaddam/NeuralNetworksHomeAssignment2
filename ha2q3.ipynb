```python
# importing all necessary dependencies
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# declaring sobel filter
sobel_x = np.array([
                    [-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]
                    ], dtype = np.float32)
sobel_y = np.array([          
                    [-1, -2, -1],
                    [0 , 0, 0],
                    [1 , 2, 1]
                    ], dtype = np.float32)
```

```python
# filtering the input image with the X and Y filters
sobelX = cv2.filter2D(image, -1, sobel_x)
sobelY = cv2.filter2D(image, -1, sobel_y)
```

Task 2: Implement Max Pooling and Average Pooling


```python
# function to perform max pooling
def max_pool(input_matrix, pool_size=2, stride=2):
    # getting height and width of input matrix
    h,w = input_matrix.shape
    # calculating output height and width
    op_height= (h-pool_size) // stride + 1
    op_weight= (w-pool_size) // stride + 1
    
    
    op_mat = np.zeros((op_height, op_weight))
    for i in range(0, h - pool_size + 1, stride):
        for j in range(0, w - pool_size + 1, stride):
            # extracting the region and getting max value of the region with the necessary stride
            region = input_matrix[i: i + pool_size, j: j+ pool_size]
            op_mat[i // stride][j // stride] = np.max(region)
            
    return op_mat
```

```python
def avg_pool(input_matrix, pool_size=2, stride=2):
    # getting height and width of input matrix
    h,w = input_matrix.shape
    op_height= (h-pool_size) // stride + 1
    op_weight= (w-pool_size) // stride + 1
   # calculating output height and width
    op_mat = np.zeros((op_height, op_weight))
    for i in range(0, h - pool_size + 1, stride):
        for j in range(0, w - pool_size + 1, stride):
             # extracting the region and getting average value of the region with the necessary stride
            region = input_matrix[i: i + pool_size, j: j+ pool_size]
            op_mat[i // stride][j // stride] = np.mean(region)
            
    return op_mat
```

```python
# declaring a 4X4 input matrix of random integersbetween 1 to 10
input_matrix = np.random.randint(1,10, (4,4))

max_mat = max_pool(input_matrix, pool_size = 2, stride = 2)
avg_mat = avg_pool(input_matrix, pool_size = 2, stride = 2)

print(f"Original Matrix: \n{input_matrix}\n")
print(f"Max Pooled Matrix: \n{max_mat}\n")
print(f"Average Pooled Matrix: \n{avg_mat}")



```
